The data was obtained from http://www.worlddev.xyz/, which allow to visualize current data from the World Bank in a way that is more accessible. 

First, to performed this analysis I chose Economy and Growth (GDP per capita (cureent international $)) and Education (labor force total) as my indicators. ## 1.Importing the data

```python
import pandas as pd 
import plotly.express as px #visualization
```

```python
df =pd.read_csv("wdi_data14countries.csv") # adding the data to be analyzed 
df.shape #This data include only the countries discussed above
```

## 2. Acessing and filtering the data

```python
df.head() #give us an idea of what type of data is included and the name of our columns
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True) # You can observed above that there is a column title unnamed
df.sample() #let's drop that column and observed the data without that column
```

## 3. Indicators

```python
df.info() # allow us to observed the columns name and the type of data in that column
```

### Indicators

* Year-> In this case the years are from 2010-2019


* SL.TFL.TOTL.IN-> Labor force total


* NY.GDP.PCAP.PP.CD-> GDP per capita, PPP (current international $)


* SL.TLF.TOTL.FE.ZS-> Labor force, female (% of total labor force)


* Country Code-> Three letter code that indentifies the counrty exp. USA


* Country Name-> Full name of the Country exp. Algeria


* Region-> Divide countries by East Asia & Pacific, North America, Middle East & North Africa, Europe & Central Asia, South Asia, Sub-Saharan Africa and Latin America & Caribbean 


* Income Group-> Does the country have low, low middle, upper middle, or high income


* Lending Type-> International Bank for Reconstruction and Developmen (IBRD), International Development Association (IDA) or blended clasifications. And countries that don't have information will appear as Not classified.



## 4. Bar graph to assess GDP in selected countries

```python
import plotly.io as pio
pio.templates.default="plotly_dark" #make the plot black
fig = px.bar(data_frame=df, 
             x="Country Code", y="NY.GDP.PCAP.PP.CD",  # I utilized the country code because the name were too long
             color="Country Name", #legend
             title="GDP per capita in selected countries (2010-2019)", 
             animation_frame="Year",
             labels={
                     "NY.GDP.PCAP.PP.CD": "GDP per capita, PPP (current international $)" # change title to be same as it in the site
             },
             )
fig.show()
```

```python
fig.write_html("bargraphGDP.html")
```

Here in this graph we can observed that United States, Germany and Canada has the higher GDP and among the lowest are Yemen and India.I run a similar raph with women in the workforce as my variable. 

## 5. Bar graph to assess female labor force in selected countries

```python
import plotly.io as pio
pio.templates.default="plotly_dark" #make the plot black
fig = px.bar(data_frame=df, 
             x="Country Code", y="SL.TLF.TOTL.FE.ZS",  # I utilized labor force %
             color="Country Name", #legend
             title="GDP per capita in selected countries (2010-2019)", 
             animation_frame="Year",
             labels={
                     "SL.TLF.TOTL.FE.ZS": "Labor force, female (% of total labor force)" # change title to be same as it in the site
             },
             )
fig.show()
```

```python
fig.write_html("bargraphlabor.html")
```

As you can see in this graph Iran, Yemen,Algeria and Saudi Arabia are amoung the counrties in this list with less female workforce. I used a buble graph to ilustrate the relation between GDP and women workforce.

## 6. Bubble graph to assess the relation between GDP and female in the workforce for the selected countries

```python
px.scatter(df, x="NY.GDP.PCAP.PP.CD", y="SL.TLF.TOTL.FE.ZS", animation_frame="Year", animation_group="Country Name",
           size="SL.TLF.TOTL.FE.ZS", color="Country Name", hover_name="Country Name",labels={
                     "SL.TLF.TOTL.FE.ZS": "Labor force, female (% of total labor force)",
                      "NY.GDP.PCAP.PP.CD": "GDP per capita, PPP (current international $)", # change title to be same as it in the site
             }, )
```

```python
fig.write_html("bubble14.html")
```

In this graph we can observed that New Zealand have a high women workforce than many countries. Nevertheless, United States and China are the countries with higher women workforce. We know that China increase in female workforce could be due to their one child policy and perhaps more favorable work evironments than other Asian countries in this list. Nevertheless we observed that Yemen, India and Iran are among the lowest in women workforce and the ones with lower GDP. With this graph it appears that with exception of Saudi Arabia, more women in the workforce correlates with more GDP. To test this hypothesis I run the same variables including all countries with data available.
